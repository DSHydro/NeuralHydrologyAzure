```python
from pathlib import Path
import pandas as pd
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
CAMELS_UY_16=pd.read_excel(Path('./caudales_filtrados/Bequelo - caudal.xls'))
CAMELS_UY_16.head()
```

```python
CAMELS_UY_16['Fecha'] = pd.to_datetime(CAMELS_UY_16['Fecha'], dayfirst=True)
CAMELS_UY_16.head()
```

```python
start_date = pd.Timestamp('1989-10-01')
end_date = pd.Timestamp('1999-09-30')

CAMELS_UY_16 = CAMELS_UY_16[(CAMELS_UY_16['Fecha'] >= start_date) & (CAMELS_UY_16['Fecha'] <= end_date)]
CAMELS_UY_16.head()
```

```python
plt.plot(CAMELS_UY_16['Fecha'], CAMELS_UY_16['Valor'])
plt.title('CAMELS_UY_16 (1989-1999)')
plt.xlabel('Date')
plt.ylabel('streamflow (m3/s)')
```

```python
start_date_nan= pd.Timestamp('1994-10-01')
end_date_nan = pd.Timestamp('1998-09-30')
CAMELS_UY_16.loc[(CAMELS_UY_16['Fecha'] >= start_date_nan) & (CAMELS_UY_16['Fecha'] <= end_date_nan), 'Valor'] = np.nan
```

```python
plt.plot(CAMELS_UY_16['Fecha'], CAMELS_UY_16['Valor'])
plt.title('CAMELS_UY_16 (1989-1999) - Corrected')
plt.xlabel('Date')
plt.ylabel('streamflow (m3/s)')
```
