This notebook creates time series of precipitation differences between the original CARAVAN dataset and the corrected CARAVAN precipitation product.

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
basin="CAMELS_UY_6"
```

```python
gauge_precip_dir=f"../preparing_data/filtered_data/time_series/{basin}.nc"
dm_precip_dir=f"./filtered_date_precip_dm/time_series/{basin}.nc"
loci_precip_dir=f"./filtered_date_precip_loci/time_series/{basin}.nc"
ls_precip_dir=f"./filtered_date_precip_ls/time_series/{basin}.nc"
pt_precip_dir=f"./filtered_date_precip_pt/time_series/{basin}.nc"
```

```python
gauge_precip=xr.open_dataset(gauge_precip_dir).prcp_mm_day
dm_precip=xr.open_dataset(dm_precip_dir).prcp_mm_day
loci_precip=xr.open_dataset(loci_precip_dir).prcp_mm_day
ls_precip=xr.open_dataset(ls_precip_dir).prcp_mm_day
pt_precip=xr.open_dataset(pt_precip_dir).prcp_mm_day
```

```python
start_date = pd.Timestamp("1989-10-01")
end_date   = pd.Timestamp("2008-09-30")
```

```python
gauge_precip=gauge_precip.sel(date=slice(start_date, end_date))
dm_precip=dm_precip.sel(date=slice(start_date, end_date))
loci_precip=loci_precip.sel(date=slice(start_date, end_date))
ls_precip=ls_precip.sel(date=slice(start_date, end_date))
pt_precip=pt_precip.sel(date=slice(start_date, end_date))
```

```python
gauge_precip
```

```python
difference_ls = gauge_precip - ls_precip
difference_ls.plot()
plt.title(f"{basin}: Gauge - LS Precipitation Difference")
plt.xlabel("Date")
plt.ylabel("Precipitation Difference (mm/day)")
plt.ylim(-100, 20)
```

```python
difference_loci = gauge_precip - loci_precip
difference_loci.plot()
plt.title(f"{basin}: Gauge - LOCI Precipitation Difference")
plt.xlabel("Date")
plt.ylabel("Precipitation Difference (mm/day)")
plt.ylim(-100, 20)
```

```python
difference_pt = gauge_precip - pt_precip
difference_pt.plot()
plt.title(f"{basin}: Gauge - PT Precipitation Difference")
plt.xlabel("Date")
plt.ylabel("Precipitation Difference (mm/day)")
plt.ylim(-100, 20)
```

```python
difference_dm = gauge_precip - dm_precip
difference_dm.plot()
plt.title(f"{basin}: Gauge - DM Precipitation Difference")
plt.xlabel("Date")
plt.ylabel("Precipitation Difference (mm/day)")
plt.ylim(-100, 20)
```
